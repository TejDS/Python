# Write a program to Create unique list from list having Duplicate  elements

```python
dup_list = [4,5,5,6,7,7]
uniq_list = []

for elem in dup_list:
    if elem not in uniq_list:
        uniq_list.append(elem)
```

```python
# Union
set_a = {2,3,4,5,6}
set_b = {2,5,7,9,10}

set_c = set_a.union(set_b)

print(f'set_a: {set_a}')
print(f'set_b: {set_b}')
print(f'set_c: {set_c}')
```

```python
# Intersection

set_a = {2,3,4,5,6}
set_b = {2,5,7,9,10}

set_c = set_a.intersection(set_b)

print(f'set_a: {set_a}')
print(f'set_b: {set_b}')
print(f'set_c: {set_c}')
```

```python
# Intersection Update
set_a1 = {2,3,4,5,6}
set_b1 = {2,5,7,9,10}

set_b1.intersection_update(set_a1)

print(set_b1)
print(set_a1)
```

```python
# Difference

set_a2 = {2,3,4,5,6}
set_b2 = {2,5,7,9,10}

set_c2 = set_a2.difference(set_b2)

print(set_c2)
```

```python
# Difference Update

set_a2 = {2,3,4,5,6}
set_b2 = {2,5,7,9,10}

set_a2.difference_update(set_b2)

print(set_a2)
```

```python
# Symmetric Difference 
set_a = {2,3,4,5,6}
set_b = {2,5,7,9,10}

set_c = set_a.symmetric_difference(set_b)

print(set_c)
```

```python
# Symmetric Difference Update
set_a4 = {2,3,4,5,6}
set_b4 = {2,5,7,9,10}

set_a4.symmetric_difference_update(set_b4)

print(set_a4)
```

```python
# Disjoint

set_a = {2,3,4,5,6}
set_b = {2,5,7,9,10}

set_a.isdisjoint(set_b)
```

```python
# Superset

set_a = {2,3,4,5,6}
set_b = {2,5,7,9,10}

set_c = set_a.union(set_b)

set_c.issuperset(set_a)
```

```python
# subset

set_a.issubset(set_c)
```

```python
set_a.issubset(set_b)
```

# Strings

It is  a string is a sequence of Unicode characters. Unicode was introduced to include every character in all languages and bring uniformity in encoding

```python
# Create a string 

a = ''
b =""
c=''''''

# using single quotes

string1 = 'Hello World'

# using double quotes

string2 = "Hello World"

# using triple quotes

string3 = '''Hello World'''

string4 = '''Hello World..!
                This is Ravi Teja'''


print(f'string1: {string1}')
print(f'string2: {string2}')
print(f'string3: {string3}')
print(f'string4: {string4}')
print(type(string4))
```

```python
# len of strings

len(string3)
```

```python
# Accessing Strings in python by using indexes and index slicing

print(string3[0]) # H
print(string3[0:5]) # Hello
print(string3[0:-1]) # Hello Worl
```

```python
# immutability of strings

string3[0] = 'L'
```

```python
# deleting strings
del string3[0]
```

```python
del string3
```

```python
print(string3)
```

```python
# Concatenate String using '+'

string1 = 'Hello '
string2 = 'World'

print(f'Concatenated String : {string1+string2}')


# Repeat String using '*'

print(string1 * 3 )
```

```python
# Writing 2 literals together

'Hello ''World''!'
```

```python
# String membership test

'el' in 'Hello World'
```

```python
'el' not in 'Hello World'
```

```python
# enumerate

string5 = 'Hello World'

enm_str = list(enumerate(string5))

print(enm_str)
```

```python
# Escape

String6 = "He said, "What's there?""
print(string6)
```

```python
# using triple quotes

print('''He said, "What's there?"''')

# using single quotes

print('He said, "What\'s there?"')


# using double quotes

print("He said, \"What's there?\"")
```

```python
# Lower
string_up = 'HELLO'
print(string_up.lower())
# Upper
string_low = 'hello'
print(string_low.upper())

# split

string_sen = 'Hi there this is Ravi Teja'

print(string_sen.split())

# join

print(' '.join(['Hi', 'there', 'this', 'is', 'Ravi', 'Teja']))


# find

print(string_sen.find('T'))


# replace

print(string_sen.replace('Hi','Hello'))

```

# Format

```python
num = 10.444444444444444444
string1 = 'My number is'

print(string1,'{0:.2f}'.format(num))
```

```python
Names = '{},{} and {}'.format('Leo','Rony','Nemo')
print(Names)
```

```python
Names = '{1},{2} and {0}'.format('Leo','Rony','Nemo')
print(Names)
```

```python
Names = '{h},{j} and {k}'.format(j='Leo',k='Rony',h='Nemo')
print(Names)
```

```python
print('The Binary format of {0} is {0:b}'.format(10))
```

```python
print('The exponential format of {0} is {0:e}'.format(10))
```
